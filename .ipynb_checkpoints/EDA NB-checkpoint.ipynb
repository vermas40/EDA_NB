```python
def num_cat(df):
    num = df.select_dtypes(include=[np.number]).columns
    cat = list(set(df.columns) - set(num))
    
    datatypes = {
        'numeric' : num,
        'categorical' : cat
    }
    
    return datatypes
```

```python
def uni_bi_numeric(df,data_types,lower_threshold = 0.8):
    print(df[data_types['numeric']].describe())
    df = df[data_types['numeric']]\
        .corr()\
        .unstack()\
        .to_frame()
    
    
    df = df \
        .loc[(df[0] > lower_threshold) & (df[0] < 1),]\
        .sort_values(by=0)\
        .drop_duplicates(subset = 0,keep = 'first')\
        .reset_index()
    
    df['rower'] = range(1,len(df)+1)
    
    return (df)    
    
```

```python
def plot_graph(main_dataset,figure_obj,x,type_of_graph,plotting_dataset = [1]):
    
    if type_of_graph == 'scatter':
        ax = figure_obj.add_subplot(len(plotting_dataset),1,x[2])
        ax.scatter(main_dataset[x[0]],main_dataset[x[1]])
    elif type_of_graph == 'line':
        ax = figure_obj.add_subplot(len(plotting_dataset),1,1)
        ax.plot(main_dataset[x[0]],main_dataset[x[1]])
    plt.xlabel(x[0])
    plt.ylabel(x[1])
```

```python
data_cache = num_cat(dataset)
```

```python
plot_dataset = uni_bi_numeric(dataset,data_cache,0.1)
```

```python
#getting the number of NAs and NaNs in pandas dataframe
dataset.fillna(0,inplace=True)
dataset.isnull().sum()
```

```python
figure = plt.figure(figsize = (10,5*len(plot_dataset)))
plot_dataset.loc[:,['level_0','level_1','rower']].apply(lambda x_send: plot_graph(dataset,figure,x_send,'scatter',plot_dataset),axis=1)
```

```python
#it works with 20 nearest neighbors
clf = LocalOutlierFactor(n_neighbors=20)
clf.fit_predict(dataset[data_cache['numeric']])
X_scores = clf.negative_outlier_factor_
```

```python
dataset = dataset.iloc[~dataset.index.isin(np.where(np.absolute(np.around(X_scores,decimals = 0)) != 1))]
```
